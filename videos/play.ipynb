```python
import cv2
from IPython.display import clear_output, Image, display
import time

cap = cv2.VideoCapture("/home/nhotin/work/FLAPPY_BIRD_RL/videos/train_step_0010000.mp4")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Chuyển sang RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    # Lưu tạm frame thành file ảnh
    cv2.imwrite("frame.jpg", frame)
    clear_output(wait=True)
    display(Image(filename="frame.jpg"))
    time.sleep(0.03)  # tốc độ hiển thị khung hình
cap.release()
```
